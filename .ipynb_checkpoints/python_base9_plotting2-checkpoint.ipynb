# bar 그래프 그리기
- bar는 어떤 특징이 갖는 값들의 빈도수를 표현할 때 유용하다.
- hist()와는 다르게 특징의 값을 정확하게 표현하고 개수를 나타낸다.
- 범주형 특징에서 사용하기 용이하다.
- 따라서 기본적으로 X축으로 특징의 값들이 나열되어야 함으로 pd.value_counts()를 많이 사용한다.
- 물론 x축의 tick을 사용자정의로 제시해도 된다.

```python
import matplotlib.pyplot as plt

df = pd.read_csv('occupancy.csv', index_col='id')
df['Occupancy'] = df['Occupancy'].map(str)
occupied = df['Occupancy'].value_counts()
occupied
```

```python
%matplotlib inline

plt.bar(x=occupied.index, height=occupied, color='c', edgecolor='yellow')
plt.show()
```

# boxplot 그래프 그리기
- boxplot은 최소값, 중앙값, 최대값, 4분위수를 표현한다.
- 사분위수란 자료값을 순서대로 나열하고, Q1(25%), Q2(50%), Q3(75%)로 4등분 하는 수의 값.
- 굳이 표현하자면, 최대값이 Q4(4-4분위수)에 해당한다. 그런데 잘 언급 안해!!
- IQR: InterQuaterRange(4분위수 범위, 4분위 편차)
- 평균은 굳이 사용하지 않으며 필요하다면 꺼내서 보여줄 순 있다.
- 이상치 판단
    - Q1 - (1.5 * IQR)
    - Q3 + (1.5 * IQR)


```python
df.head()
```

```python
plt.boxplot(x=df.Temperature, notch=True, showmeans=True, 
           meanline=True, vert=True, labels=["온도"])
```

```python
# 두 개 이상의 boxplot 그리기
plt.boxplot(x=[df.Temperature, df.Humidity], notch=True, showmeans=True, 
           meanline=True, vert=False, labels=["온도", '습도'])
```

# hist 그래프 그리기

# 산점도 그래프 그리기
- 두 변수 사이의 상관관계를 파악하기에 유용함.
- 점의 모양을 바꾸어 선 모양으로 표현할 수 있음.
- plt에서는 두 가지 방법으로 산점도를 그릴 수 있다.
    - plt.plot(marker='*', linestyle="") # y만 넣어줘도 크기에 맞춰 x는 알아서 나옴.
    - plt.scatter(x=, y=) # x, y 값을 넣어줘야 해.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('APT.csv', encoding='cp949')
df.head()
```

```python
plt.plot(df.Price, marker='v', linestyle='', color='m')
```

```python
plt.scatter(np.arange(len(df)), df.Price, color='b')
```

# 상관관계 함수
- pd.corr()
- corr()의 옵션으로 method='pearson', 'kendall', 'spearman'이 있다. 기본은 'pearson'
- pd.corrwith(특정하나의 컬럼 시리즈)
    - 특정 하나의 컬럼 시리즈와 나머지와의 상관관계 파악할 때 사용
- 통계적으로 0.3 이상이면 약간의 상관관계가 있다고 하는데 체감하기는 어려움.
- 보수적으로는 0.7 이상이면 상관관계가 있다고 본다.

```python
import matplotlib.pyplot as plt

df = pd.read_csv('occupancy.csv', index_col='id')
df.corr()
```

```python
df.corrwith(df.Light)
```

```python
plt.scatter(df.Light, df.Temperature)
```

# seabon
- Matplotlib을 기반으로 다양한 색상 테마와 기능을 추가한 시각화 패키지
- 자주 사용이 되나, Matplotlib과는 사용법이 조금 다르다
- 기본적인 시각화 기능은 Matplotlib 패키지에 의존
- 통계 기능은 Statsmodels 패키지에 의존

- <u>**sns.set()**</u>
    - 그래프의 전반적인 스타일을 Seaborn 기본 스타일로 변경
    - 주요 옵션
        - context : 라벨 스타일(paper, notebook, talk, poster)
        - style : 배경 스타일(darkgrid, whitegrid, dark, white, ticks)
            - https://python-graph-gallery.com/104-seaborn-themes
        - palette : 색상 스타일(deep, muted, pastel, bright, dark, colorblind)
            - https://seaborn.pydata.org/tutorial/color_palettes.html
        - font_scale : 폰트 사이즈

# seaborn의 그래프 함수들
- <u>**아래 함수들은 많이 헷갈리니까 help(sns.boxplot) 등을 이용하자.**</u>
- barplot(x=, height=)
    - X는 이산형 또는 범주형
    - height는 count가 될 수도 있고, 다른 연속형 데이터일 수 있는데, 연속형일 때는 같은 이산형 또는 범주형에서의 평균값이 계산되어 들어간다.
- boxplot(x=)
- distplot(x=)
    - histogram과 kde를 같이 보여줌.
    - kde=False 옵션을 이용하여 확률밀도함수 그래프를 제외시킬 수 있다.
    - 데이터의 분포를 보고 kde를 계산한 것. 데이터 수가 적을 때 유용함.

## seaborn barplot 그래프 그리기

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = '/Users/jsha/gjai/python_basic/pytest_basic/'
df = pd.read_csv(path+'APT.csv', encoding='euc-kr')
df.head()
df.dropna(inplace=True)
df
```

```python
sns.set()
sns.barplot(x=df.id, y=df.Price)
plt.show()
```

## seaborn boxplot 그래프 그리기

```python
sns.set(context="notebook", style="white", palette="pastel")

sns.boxplot(x=df.Price)
plt.xlabel("APT Price")
plt.show()
```

```python
sns.boxplot(y=df.Price, orient='v')
plt.ylabel("APT Price")
plt.show()
```

## seaborn distplot 그래프 그리기

```python
sns.set(palette='dark')

sns.distplot(x=df.Price, color='red')
plt.xlabel("APT Price")
plt.show()
```

```python
sns.distplot(x=df.Price, color='red', kde=False)
plt.xlabel("APT Price")
plt.show()
```

## seaborn relplot 그래프 그리기
- option kind='line'을 써서 선 그래프를 그릴 수 있음.
- sns.scatterplot()과 무슨 차이인지 잘 모르겠음.

```python
sns.set(context='talk')

sns.relplot(x="Price", y="FAR", data=df, color='green')
plt.xlabel("APT Price")
plt.show()
```

```python
sns.scatterplot(x="Price", y="FAR", data=df, color='green')
plt.xlabel("APT Price")
plt.show()
```

```python
sns.relplot(x="Price", y="FAR", data=df, color='green', kind='line')
plt.xlabel("APT Price")
plt.show()
```

## seaborn lmplot 그래프 그리기 - 회귀선

```python
sns.lmplot(x='FAR', y='Price', data=df)
plt.show()
```
