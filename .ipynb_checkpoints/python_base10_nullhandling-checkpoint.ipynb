# 이상치 다뤄보기
1. 이상치
    - outlier
    - 이상치란 데이터에서 다른 값보다 아주 작거나 큰 수
    - 특이하고 이상한 데이터
    - 모형 구축에 있어 이상치는 그 빈도에 비해 모델 결과에 큰 영향을 미쳐 모수 추정에 어려움을 줄 수 있으므로 반드시 관심을 기울여야 함.
    - 그러나 <u>**이상치가 항상 의미 없는 값은 아니므로**</u>, 본 데이터에 대한 전문가가 반드시 확인해야 함. 새로운 인사이트가 나올 수도 있음.
    
    
2. 이상치 탐색
    - **통계기법으로 정밀하게 이상치를 찾아야 한다.**
    - describe()를 이용해서 전체 컬럼별 통계분석을 한다.
    - 가장 중요한 min, max를 보고, 그 다음 mean을 살펴보자.
    - boxplot()으로 한 눈에 컴럼별 이상치를 파악해 본다. 
    - 그러나 이 boxplo()을 너무 맹신하진 마. 박스를 벗어나 아무 넓게 분포하면 이는 문제가 되며 자세히 살펴야 한다. 낮은 이상치, 높은 이상치
    - 근데 웃긴게 **결국 이상치를 제거하고 박스플랏을 하면 또 이상치로 판단할 수 있는게 나옴.** 결국 엄청나게 큰 수라고 해서 이상치라고 접근하지 말고 그 값이 어떻게 나왔을지에 대해 유추해 보는 게 필요해 보임.
    
    
3. 이상치 처리 방법
 - 제거
     - 오타, 오류, 비상식적 응답(반응)인 경우
 - 치환
     - 평균, 최빈값, 중앙값, 예측값으로 치환
     - 신뢰도의 문제가 발생
 - 분리
     - 독립변수가 충분히 세분화 되지 않았을 때 발생할 수 있음.
     - 예를 들어, 2022년 데이터가 2021년에 포함되었다던가 어떤 기준에 의해 나눴어야 했는데 그러지 못한 경우

## 사분위수를 이용한 이상치 제거

```python
q1 = df.describe().loc['25%', 'crim']
q3 = df.describe().loc['75%', 'crim']

print(f'1사분위수: {q1:.2f}', f'1사분위수: {q3:.2f}')
low_outlier = q1 - 1.5*(q3 - q1)
high_outlier = q3 + 1.5*(q3 - q1)
print(f'낮은 이상치: {low_outlier:.2f}', f'높은 이상치: {high_outlier:.2f}')
```

```python
# 이상치 찾고 df 만들기
outlier_df = df[(df['crim'] > high_outlier) | (df['crim'] < low_outlier)]
outlier_df

# 이상치를 제외한 df 만들기
new_df = df[(df['crim'] < high_outlier) & (df['crim'] > low_outlier)]
new_df.describe()
```

```python
new_df.boxplot(column='crim')
```
