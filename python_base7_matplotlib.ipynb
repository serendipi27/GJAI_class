# 연습문제 1

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import rc

rc('font', family='AppleGothic')

path = '/Users/jsha/gjai/python_basic/pytest_basic/'
os.chdir(path)

df = pd.read_csv(path+'electricity.csv', encoding='cp949')
df.head()
```

```python
part_arr = np.unique(df['Part'])
part_arr
df_dict = {}
for i in range(len(part_arr)):
    df_dict[part_arr[i]] = df[df['Part']==part_arr[i]].sort_values('Year')
df_dict

for key, value in df_dict.items():
    plt.plot(value.Year, value.Usage, label = key)
plt.legend()
plt.title('Electricity Usage')
plt.xlabel('Year')
plt.ylabel('Usage')
plt.show()
```

```python
# seaborn을 이용해 한방에 해결하기
import seaborn as sns
sns.lineplot(x='Year', y='Usage', data=df, hue='Part')
```

# 산점도 그리기
 - plt.scatter()
 - 사이즈(s=100), 투명도(alpha=0.3), 컬러(c='indigo')

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

plt.style.use('seaborn-whitegrid')

weight = np.random.randint(150, 180, size=50)
height = np.random.randint(50, 90, size=50)

plt.scatter(sorted(weight), sorted(height), color='c')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

plt.style.use('seaborn-whitegrid')

weight = np.random.randint(150, 180, size=50)
height = np.random.randint(50, 90, size=50)

plt.scatter(sorted(weight), sorted(height), 
            color='indigo', s=100, alpha=0.6)
plt.title('Relation of Weight and Height')
plt.xlabel('weight')
plt.ylabel('height')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

plt.style.use('seaborn-whitegrid')

height_man = np.random.randint(150, 180, size=50)
height_women = np.random.randint(140, 170, size=50)
weight = np.random.randint(50, 90, size=50)


plt.scatter(sorted(weight), sorted(height_man), 
            color='b', s=50, alpha=0.6, label='Man')
plt.scatter(sorted(weight), sorted(height_women), 
            color='indigo', s=100, alpha=0.6, label='Women')
plt.title('Relation of Weight and Height')
plt.xlabel('weight')
plt.ylabel('height')
plt.legend()
plt.show()
```

# 연습 문제2
 - heightweght.csv 파일 이용
 - 남녀의 키차이를 산점도로 그려라.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('font', family='AppleGothic')

path = '/Users/jsha/gjai/python_basic/pytest_basic/'
df = pd.read_csv(path+'heightweight.csv')
df.describe()
df.head()

male_df = df[df.sex == 'm']
female_df = df[~(df.sex == 'm')]

df.columns

plt.scatter(male_df.heightIn, male_df.weightLb, s=100, c='c', alpha=0.6, label='male')
plt.scatter(female_df.heightIn, female_df.weightLb, s=50, c='b', alpha=0.2, label='female')
plt.title('남-여 키-몸무게 비교')
plt.xlabel('height')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(data=df, x=df.heightIn, y=df.weightLb, hue=df.sex)
```

# 연습 문제3
- iris.csv 파일은 붓꽃의 하위 품종인 setosa, versicolor, virginica 의 특징을
기록한 파일임.
- 품종의 특징이 잘 드러나도록 두 개 컬럼을 선택하여 산점도를 그려라.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('font', family='AppleGothic')

path = '/Users/jsha/gjai/python_basic/pytest_basic/'
df = pd.read_csv(path+'iris.csv')
df.columns

fit, ax = plt.subplots(4,4)
i = 0
for col1 in df.columns:
    for col2 in df.olumns:
        ax[i].scatter(df[col1], df[col2])
        ax[i].set_xlabel(col1)
        ax[i].set_ylabel(col2)
        
        i += 1
```
